```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the file
data = pd.read_csv('CO2 Emissions_Canada.csv')
data
```

```python
#finding is there any null values
data.isnull().sum()
```

```python
#Checking duplicate values
data[data.duplicated()]
```

```python
#Removing duplicate values
data.duplicated().value_counts()
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
#Reducing the data to 100 rows
emi=data.iloc[0:100,0:12]
```

```python
#Displaying all data's in 100 rows
emi
```

```python
#Displaying basic statistical functions
emi.describe()
```

```python
#Plot a bar chart where the X axis represent Cylinders and Y axis represent Combined fuel consumption
plt.bar(emi['Cylinders'],emi['Fuel Consumption Comb (L/100 km)'])
plt.xlabel("Cylinders")
plt.ylabel("Fuel Consumption Comb (L/100 km)")
plt.title("Fuel Consumption")
#Display the plot
plt.show()
```

```python
#Create a scatter plot with Engine size and CO2 emissions
plt.scatter(emi['Engine Size(L)'],emi['CO2 Emissions(g/km)'])
plt.xlabel("Engine Size(L)")
plt.ylabel("CO2 Emissions(g/km)")
plt.title("Relation Between Engine Size and CO2 Emissions")
plt.show()
```

```python
#Checking the correlation
co=data[['Engine Size(L)', 'Cylinders', 'Fuel Consumption City (L/100 km)', 'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)','Fuel Consumption Comb (mpg)','CO2 Emissions(g/km)']].corr()
co
```

```python
#Calculating the median for cylinders
media=data['Cylinders'].median()
media
```

```python
#Calculating the standard deviation for engine size
standard=data['Engine Size(L)'].std()
standard
```

```python
#Calculating kurtosis for fuel consumption hwy
kurtos=data['Fuel Consumption Hwy (L/100 km)'].kurtosis()
kurtos
```

```python
#Calculating skew value of fuel consumption comb 
ske=data['Fuel Consumption Comb (L/100 km)'].skew()
ske
```

```python
numeric_df = data.select_dtypes(include='number')
```

```python
#Creating the correlation matrix
plt.figure(dpi=144)
#data_numeric = data.apply(pd.to_numeric, errors='coerce')
plt.figure(figsize=(20,20))
plt.title('correlation Matrix',fontsize=25)
sns.heatmap(numeric_df.corr(),cmap='RdBu', annot=True)
plt.title("Correlation",size=20)
plt.show
```
